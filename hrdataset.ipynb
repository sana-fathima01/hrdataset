```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
hr=pd.read_csv('C:/Users/Admin/Downloads/HRDataset.csv')
print(hr)
```

```python
print("\n",hr.isnull().sum())
```

```python
hr['DateofTermination']=hr['DateofTermination'].fillna(hr['DateofTermination'].mode()[0])
hr['ManagerID']=hr['ManagerID'].fillna(hr['ManagerID'].mode()[0])
print(hr)
print("\n",hr.isnull().sum())
```

```python
plt.figure(figsize=(19,6))
sns.lineplot(data=hr,x='EmpID',y='MaritalStatusID',hue='Department',marker='o')
plt.title('EmployeeID vs maritalStatusID')
plt.xlabel('EmpID')
plt.ylabel('MaritalStatusID ')
plt.grid(True)
plt.show()
```

```python
avg_salary=hr.groupby('Department')['Salary'].mean().reset_index()
plt.figure(figsize=(10,5))
sns.barplot(data=avg_salary,x='Department',y='Salary',palette='viridis')
plt.title('Average salary by department')
plt.xlabel('Department')
plt.ylabel('Average salary')
plt.show()
```

```python
plt.figure(figsize=(6,20))
sns.scatterplot(data=hr,x='EmpID',y='DateofTermination',hue='Department',palette='Set1')
plt.title('EmpID vs DateofTermination ')
plt.xlabel('EmpID')
plt.ylabel('DateofTermination')
plt.legend(title='Department')
plt.show()
```

```python
plt.figure(figsize=(20,6))
sns.histplot(data=hr,x='RecruitmentSource',hue='Department',kde=True,palette='Set2',bins=15)
plt.title('Distribution of RecruitmentSource by Department')
plt.xlabel('RecruitmentSource')
plt.ylabel('frequency')
plt.show()
```

```python
numeric=hr.select_dtypes(include=['float64','int64'])
corr_matrix=numeric.corr()
plt.figure(figsize=(25,6))
sns.heatmap(corr_matrix,annot=True,cmap='coolwarm',fmt='2f',linewidth=0.5)
plt.title('correlation heatmap of human resourse dataset')
plt.show()
```
